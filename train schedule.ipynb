```python
df['departure'] = pd.to_datetime(df['departure'], errors='coerce')

plt.figure()
df['departure'].dt.hour.dropna().plot(kind='hist', bins=24)
plt.title("Departure Time Distribution (Hours)")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trains")
plt.show()
```

```python
plt.figure()
df['train_name'].str.extract('(Express|Passenger|MEMU|MMTS)')[0] \
    .value_counts().plot(kind='bar')
plt.title("Train Type Distribution")
plt.xlabel("Train Type")
plt.ylabel("Count")
plt.show()
```

```python
daily_trains = df.groupby('day')['train_number'].nunique()

plt.figure()
daily_trains.plot(kind='line', marker='o')
plt.title("Unique Trains per Day")
plt.xlabel("Day")
plt.ylabel("Number of Unique Trains")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression


X = pd.DataFrame({'day': [1,2,3,4,5]})
y = [47154, 56044, 19269, 58205, 63542]

model = LinearRegression()
model.fit(X, y)


new_data = pd.DataFrame({'day': [6]})
prediction = model.predict(new_data)

print("Prediction:", prediction)
print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_actual = [1,2,3,4,5]
y_pred   = [47154,56044,19269,58205,63542]

mae = mean_absolute_error(y_actual, y_pred)
mse = mean_squared_error(y_actual, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_actual, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("RÂ² Score:", r2)
```
